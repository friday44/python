Решение дается в общем виде, что нас не устраивает для дальнейшей работы.

Выполняем все поставленные задачи для функции f(x) = -12*x**4 - 18*x**3 + 5*x**2 + 10*x - 30

1. Определить корни

```python
f = - 12*x**4 - 18*x**3 + 5*x**2 + 10*x + 30
from sympy import solveset, N
list(map(N, solveset(f, x)))
```

4. Построить график

```python
from sympy.plotting import plot
plot(f, (x, -3, 2))
```

Если найти производную, потом её корни, потом вычислить в этих точках функцию, то мы получим экстремумы

Производная

```python
f_diff = sympy.diff(f, x)
f_diff
```

```python
list(map(N, solveset(f_diff, x)))
```

```python
plot(f_diff, (x, -1.5, 1))
```

```python
х_extr = [0.43, -0.41, -1.15]
y_extr = list(map(lambda х:f_diff, х_extr))
y_extr
```
