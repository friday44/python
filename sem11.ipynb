3. Найти интервалы, на которых функция убывает

```python
from sympy import is_decreasing, is_increasing, oo
is_increasing(5*x**2 + 10*x - 30, Interval.open(-1, oo))
```

```python
is_decreasing(5*x**2 + 10*x - 30, Interval.open(-oo, -1))
```

```python
на участке (-oo, -1) функция убывает; на участке (-1, oo) функция возрастает
```
